##Authenticate Colab with GCP

```python
from google.colab import auth
auth.authenticate_user()
```

##Set project

```python
project_id = "olist-analytics-493621"  # replace with your real project ID

import subprocess
subprocess.run(["gcloud", "config", "set", "project", project_id])

print(f"Project set to: {project_id}")
```

##Test BigQuery connection

```python
from google.cloud import bigquery

client = bigquery.Client(project=project_id)

# Run a simple test query
query = """
    SELECT
        schema_name
    FROM
        INFORMATION_SCHEMA.SCHEMATA
"""

result = client.query(query).to_dataframe()
print("Connected! Datasets in your project:")
print(result)
```

##Test GCS connection

```python
from google.cloud import storage

storage_client = storage.Client(project=project_id)

buckets = list(storage_client.list_buckets())
print("Connected! Buckets in your project:")
for bucket in buckets:
    print(f"  gs://{bucket.name}/")
```
